# 데이터 셋

## 1. MNIST Dataset 불러오기

```python
dataset_name = "mnist"

train_dataset = LocalizationDataset(dataset_name,
                                    data_type='train',
                                    rescale_ratio=(0.5,3.0))
validation_dataset = LocalizationDataset(dataset_name,
                                    data_type='validation',
                                    rescale_ratio=(0.5,3.0))
test_dataset = LocalizationDataset(dataset_name,
                                   data_type='test',
                                   rescale_ratio=(0.5,3.0))
```

## 2. 데이터셋 확인하기

```python
# 데이터 셋에서 이미지 하나 가져오기
image, point, label = train_dataset[0]

plt.imshow(image)
plt.show()
```

### 우리의 목표

우리는 위의 이미지를 통해 2 개의 결과값을 반환해야 합니다.

* 이 숫자가 어디에 있을까? -> Localization
* 이 숫자가 그래서 얼마야? -> Classification

## 3. Generator 불러오기

```python
# Keras의 모델에 넣을 수 있는 형태로, 데이터셋을 Transform하는 generator
train_generator = DataGenerator(train_dataset)
validation_generator = DataGenerator(validation_dataset)
test_generator = DataGenerator(test_dataset)
```

# Localization Model

모델은 아래와 같은 형태로, 3가지로 구성되어 있다.

![Imgur](https://i.imgur.com/rigiCPF.png)

### 1. 케라스로 구현한 모델 가져오기

```python
# 아래로 모델을 불러올 수 있음 -> 코드를 통해 확인하길 바람 ㅠ
# 코드가 길어져서, 여기 넣으면 너무 지저분할거라 생각
model = tiny_localization_network((None,None,1), 10)

# compile 단계는 학습에 대한 Setting을 하는 단계라 생각하면 됨
model.compile(optimizer=Adam(lr=0.001),
              metrics={
                  "localization_output":iou_metric, 
                  "classification_output":accuracy_metric
              },
              loss={
                  "localization_output":smooth_l1_loss,
                  "classification_output":categorical_crossentropy
              },
              loss_weights={
                  "localization_output":1,
                  "classification_output":0.1
              })
# 모델 개괄 보기
model.summary()
```

### 2. Keras 모델 학습

```python
# Log directory 정리
if os.path.exists(LOG_DIR):
    shutil.rmtree(LOG_DIR)
os.makedirs(LOG_DIR, exist_ok=True)

# 시각화에 이용할 데이터들
val_images, [val_centers, val_labels] = validation_generator[0]

# Callback함수들, 이를 통해 우리는 Tensorboard에서 어떻게 학습되어 가는지를 확인할 수 있다.
callbacks = [
    TensorBoard(LOG_DIR),
    TensorBoardImage(val_images, val_centers, val_labels, LOG_DIR)]
```

```python
hist = model.fit_generator(train_generator,
                           epochs=20,
                           validation_data=validation_generator,
                           callbacks=callbacks)
```

> 텐서보드에 접속하면 위의 학습과정을 관찰할 수 있습니다.

## 2. 테스트 데이터셋으로 결과를 확인하기

```python
images, [true_centers, true_labels] = test_generator[0]
pred_centers, pred_labels = model.predict(images)
```

```python
height, width = images.shape[1:3]
rescaling = np.array([width,width,height,height])
pred_points = (center2minmax(pred_centers.squeeze())
               *rescaling).astype(np.int)
true_points = (center2minmax(true_centers.squeeze())
               *rescaling).astype(np.int)

results = []
for image, pred_point, pred_label, true_point, true_label in zip(
        images, pred_points, pred_labels, true_points, true_labels):
    if image.dtype == np.float:
        image = (image*255).astype(np.uint8)

    if image.ndim == 3 and image.shape[-1] == 1:
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    elif image.ndim == 2:
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    
    pred_point = pred_point.squeeze()
    true_point = true_point.squeeze()
    pred_label = pred_label.squeeze()
    true_label = true_label.squeeze()
    
    image = cv2.rectangle(image,
                          (pred_point[0],pred_point[2]),
                          (pred_point[1],pred_point[3]),(255,0,0),1)
    image = cv2.rectangle(image, 
                          (true_point[0],true_point[2]),
                          (true_point[1],true_point[3]),(0,255,0),1)
    
    text = "{}".format(pred_label.argmax())
    image = cv2.putText(image, text,(0,10),
                        cv2.FONT_HERSHEY_DUPLEX,0.3,(255,0,0))
    text = "{}".format(true_label.argmax())
    image = cv2.putText(image, text,(image.shape[1]-20,10),
                        cv2.FONT_HERSHEY_DUPLEX,0.3,(0,255,0))
    plt.imshow(image)
    plt.show()
```
